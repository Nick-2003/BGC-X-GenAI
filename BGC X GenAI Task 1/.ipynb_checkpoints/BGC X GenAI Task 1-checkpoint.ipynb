# BGC X GenAI Task 1

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data collection and loading
- Data collected under several assumptions:
    - For `Tesla`, "Net cash provided by operating activities" under "Cash Flow from Operating Activities" = "Cash Flow from Operating Activities"
    - For `Apple`, "Total net sales" = "Total Revenue".
- Encoding done to deal with leading whitespaces.
- `pct_change()` uses the current and prior data to obtain the result, so `Year` was set in ascending order.

```python
df = pd.read_csv('BGC_X_GenAI_Task_1.csv', encoding='utf-8-sig', sep='\s*,\s*', engine='python')
df = df[df['Company'].notna()] # https://stackoverflow.com/questions/13413590/how-to-drop-rows-of-pandas-dataframe-whose-value-in-a-certain-column-is-nan
display(df)
```

## Feature engineering
- Growth columns created to analyse year-over-year changes in financial metrics.
- Profit margin as measure of profitability of business.
    - Profit margin = $\frac{\text{Net Income}}{\text{Total Revenue}}$
- Return on assets as measure of business asset utilisation.
    - Return on assets = $\frac{\text{Net Income}}{\text{Total Assets}}$
- Operating cash flow ratio as measure of operational efficiency. 
    - Operating cash flow ratio = $\frac{\text{Operating Cash Flow}}{\text{Total Revenue}}$

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df['Asset Growth (%)'] = df.groupby(['Company'])['Total Assets'].pct_change() * 100
df['Liability Growth (%)'] = df.groupby(['Company'])['Total Liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby(['Company'])['Cash Flow from Operating Activities'].pct_change() * 100

df['Profit Margin'] = df['Net Income'] / df['Total Revenue']
df['Return On Assets'] = df['Net Income'] / df['Total Assets']
df['Operating Cash Flow Ratio'] = df['Cash Flow from Operating Activities'] / df['Total Revenue']
display(df)
```

## `df.info()`
- All variables except for `Company` and `Year` are numeric.
- `Company` and `Year` are meant to be categorical.

```python
display(df.info())
```

```python
# display(df.describe())
```

## Boxplots of base metrics
- Boxplots were used to allow for direct comparisons between each company within each year.
- In terms of total revenue, income, liabilities, and operating cash flow, Apple has significantly greater values than Microsoft, who in turn has significantly greater values than Tesla.
- Apple and Microsoft have similar total assets, while Tesla has significantly lower total assets than either company.

```python
numericcats = ["Total Revenue", "Net Income", "Total Assets", "Total Liabilities", "Cash Flow from Operating Activities"]
for n in numericcats: 
    df.groupby(['Year', 'Company'])[n].sum().unstack().plot(kind='bar', figsize=(10, 5))
    plt.title(str(n) + " for each company and year")
    plt.xlabel("Year")
    plt.ylabel('Amount')
    plt.show()
```

## Boxplots of growth per year for each metric
- Growth is relative to 1 year prior to the given year.
- `dropna()` is applied since there is no growth from 2020 to 2021 provided in the data.
- Tesla has significantly higher growth for each metric compared to Apple and Microsoft, but the metric growths from 2022 to 2023 are lower than the metric growths from 2021 to 2022.
- Apple's metric growths are all positive from 2021 to 2022, and negative or 0% from 2022 to 2023.
- Microsoft's growths for revenue, net income, and cash flow decrease over time, while their growth in assets and liabilities are slightly increased over time.

```python
createdcats = ["Revenue Growth (%)", "Net Income Growth (%)", "Asset Growth (%)", "Liability Growth (%)", "Cash Flow Growth (%)"]
for n in createdcats: 
    df.dropna().groupby(['Year', 'Company'])[n].sum().unstack().plot(kind='bar', figsize=(10, 5))
    plt.title(str(n) + " for each company and year")
    plt.xlabel("Year")
    plt.ylabel('Amount')
    plt.show()
```

## Boxplots of metric ratios 
- Microsoft has significantly greater profit margin and operational efficiency compared to Apple, which in turn has significantly greater profit margin and operational efficiency compared to Tesla.
- Apple has significantly greater profit margin and operational efficiency compared to Microsoft, which in turn has significantly greater profit margin and operational efficiency compared to Tesla.

```python
createdcats2 = ["Profit Margin", "Return On Assets", "Operating Cash Flow Ratio"]
for n in createdcats2: 
    df.groupby(['Year', 'Company'])[n].sum().unstack().plot(kind='bar', figsize=(10, 5))
    plt.title(str(n) + " for each company and year")
    plt.xlabel("Year")
    plt.ylabel('Amount')
    plt.show()
```

## Conclusion 
Given the growths and ratios of the provided metrics, Tesla consistently has the highest growth and the lowest cash flows among the given companies, implying potential sustainability issues and liquidity risks. Apple consistently demonstrates high cash flows, but low or even negative growths compared to the other companies, suggesting great financial stability along with lack of long-term potential. Microsoft often hits the middle ground in terms of cash flows and growths, along with often attaining relatively high metric ratios, which implies that it is relatively stable in terms of both the short term and the long term, though with likely lower returns than Apple or Tesla.
